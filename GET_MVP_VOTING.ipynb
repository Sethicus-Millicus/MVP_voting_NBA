```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from urllib.request import urlopen
import time
```

```python
years = [str(i) for i in list(range(1980, 2020))]
```

```python
for year in years:
    time.sleep(3)
    print(year)
    url = "https://www.basketball-reference.com/awards/awards_{}.html".format(year)
    html = urlopen(url)
    soup = BeautifulSoup(html)
    yr = soup.find('h1').getText().split()[0]
    headers = [th.getText() for th in soup.find('thead').findAll('th')]
    del headers[:7]
    table = soup.find('tbody')
    rows = table.findAll('tr')
    data = [[td.getText() for td in rows[i].findAll('td')]
           for i in range(len(rows))]
    mvp_results = pd.DataFrame(data, columns = headers)
    mvp_results.insert(0, 'Season', yr)
    mvp_results.to_csv('NBA_by_year_stats/{}_mvp_voting.csv'.format(year))
```

```python
url = "https://www.basketball-reference.com/awards/awards_2020.html"
html = urlopen(url)
soup = BeautifulSoup(html)
yr = soup.find('h1').getText().split()[0]
headers = [th.getText() for th in soup.find('thead').findAll('th')]
del headers[:7]
table = soup.find('tbody')
rows = table.findAll('tr')
data = [[td.getText() for td in rows[i].findAll('td')]
        for i in range(len(rows))]
mvp_results = pd.DataFrame(data, columns = headers)
mvp_results.insert(0, 'Season', yr)
mvp_results.to_csv('NBA_by_year_stats/2020_mvp_voting.csv')
```
